# **Escape Sequence Character**

```python
# now we can see the escape sequence character
a = "everything going\nbored" # it is use for start next line
print(a)
a = "everything going\"bored\"" # it is use for add "" in answer
print(a)
a = "everything going\tbored" # it is use for add space on line
print(a)
a = "everything going\abored" # it is use for add special charachter in a line
print(a)
a = "everything going\\bored" # it is use for add \ in a line
print(a)
a = "everything going\fbored" # it is use for less gap between from words
print(a)
```
