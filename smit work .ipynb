```python
# sets are mutable
set: set = {1,2,33,4,5,6}
print("set:",set)
a = {1,26,6,19}
c = {33,66,6,12}
a.add(13)
a.remove(26)
print("set","intersection",a & c)# intersection
print("set","union (sorted):",sorted (a | c))# union sorted means the number in row unsorted means number are disturb
print(a - c)# this sign is use in set for define difference
f=frozenset([1,2,89])
print(f)

```

```python
A = frozenset([1, 2, 3])
B = frozenset([3, 4, 5])

print(A | B)  # Union
print(A & B)  # Intersection
print(A - B)  #difference
a.add(5)
```
